```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_dsa = pd.read_csv('/dados_clientes.csv')
```

```python
df_dsa.head(10)
```

```python
#Resumo Estatistico
df_dsa[['idade','renda_anual','pontuacao_gastos']].describe()
```

```python
#Cria um padronizador de Dados
padronizador = StandardScaler()
```

```python
#Aplica o padronizador somente nas colunas de interesse
dados_padronizados = padronizador.fit_transform(df_dsa[['idade','renda_anual','pontuacao_gastos']])
```

```python
#Visualiza os dados
print(dados_padronizados)
```

```python
#Definimos o nmero de clusters (k)
k = 3
```

```python
#criamos o modelo de clusters k-means
kmeans = KMeans(n_clusters = k)
```

```python
#treinnamentos do modelo com os dados padronizados
kmeans.fit(dados_padronizados)
```

```python
#atribuimos os rotulos dos clusters aos clientes
df_dsa['cluster'] = kmeans.labels_
```

```python
#Exibe o resultado (10 primeiras linhas)
df_dsa.head(10)
```

```python
df_dsa.to_csv('/dados_clientes_clusterizados.csv', index=False)
```
